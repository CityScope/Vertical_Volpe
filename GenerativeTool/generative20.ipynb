## COMMON FUNCTIONS

```python
from scipy.spatial.distance import cdist

def closest_point(point, points):
    """ Find closest point from a list of points. """
    return points[cdist([point], points).argmin()]

def match_value(df, col1, x, col2):
    """ Match value x from col1 row to value in col2. """
    return df[df[col1] == x][col2].values[0]
```

## FIRST WE WILL LOOK FOR IDS FROM SOURCES

```python
points_w_landUses_residences = points_w_landUses_df[points_w_landUses_df['type_of_entity'].isin(
    ['QUAD_OCCUPANCY', 'DUAL__OCCUPANCY', 'TRIPLE_OCCUPANCY', 'SINGLE_OCCUPANCY','FAMILY_OCCUPANCY'])]
```

```python
source_nodes = pd.read_csv("source_nodes.csv")
```

```python
source_nodes['point'] = [(x, y, z) for x,y,z in zip(source_nodes['lat'], source_nodes['lon'], source_nodes['z'])]
points_w_landUses_residences['point'] = [(x, y, z) for x,y,z in zip(points_w_landUses_residences['lat'], points_w_landUses_residences['lon'], points_w_landUses_residences['z'])]
points_w_landUses_residences['closest'] = [closest_point(x, list(source_nodes['point'])) for x in points_w_landUses_residences['point']]
points_w_landUses_residences['id_closest'] = [match_value(source_nodes, 'point', x, 'id') for x in points_w_landUses_residences['closest']]
```

## NOW WE LOOK FOR IDS FROM TARGETS

```python
target_nodes = pd.read_csv("target_nodes.csv")
```

```python
points_w_landUses_targets = points_w_landUses_df[points_w_landUses_df['type_of_entity'].isin(
    ['OFFICES', 'BAR', 'RESTAURANT', 'PARKS', 'HEALTH_CARE', 'CAFE',
       'LAUNDRY', 'BANKS', 'LIBRARY', 'GROCERY_OR_SUPERMARKET',
       'HOME_GOODS_STORE', 'PROFESSIONAL_SERVICES', 'PHARMACY',
       'CONVENIENCE_STORE', 'SCHOOL', 'FURNITURE_STORE',
       'LEISURE_AND_WELLNESS'])]
```

```python
target_nodes['point'] = [(x, y, z) for x,y,z in zip(target_nodes['lat'], target_nodes['lon'], target_nodes['z'])]
points_w_landUses_targets['point'] = [(x, y, z) for x,y,z in zip(points_w_landUses_targets['lat'], points_w_landUses_targets['lon'], points_w_landUses_targets['z'])]
points_w_landUses_targets['closest'] = [closest_point(x, list(target_nodes['point'])) for x in points_w_landUses_targets['point']]
points_w_landUses_targets['id_closest'] = [match_value(target_nodes, 'point', x, 'id') for x in points_w_landUses_targets['closest']]
```

## NOW WE CREATE THE PROXIMITY MATRIX FOR THE VOXELIZED NODES

```python
#First we put a placeholder matrix
reachable_vox_matrix = np.zeros((points_w_landUses_residences.shape[0],points_w_landUses_targets.shape[0])).astype(int)
#Then we loop among elements
```

```python
for ind_r,dot_r in enumerate(points_w_landUses_residences.values):
    print("{} of {} points evaluated".format(ind_r,len(points_w_landUses_residences.values)))
    for ind_t,dot_t in enumerate(points_w_landUses_targets.values):
        residence = reachable_df[reachable_df['Origins']==dot_r[7]]
        reachable_vox_matrix[ind_r][ind_t] = residence[dot_t[7]]
        
```
